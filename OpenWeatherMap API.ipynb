# Current weather

```python
url1="https://api.openweathermap.org/data/2.5/weather?lat="+str(lat)+"&lon="+str(long)+"&units=metric&appid="+api_key
response1 = requests.get(url1)
current_weather = response1.json()
# print(current_weather)
print(json.dumps(current_weather, indent = 5))
```

```python
pressure = current_weather['main']['pressure']
print("Pressure: ",pressure)
```

```python
humidity = current_weather['main']['humidity']
print("Humidity: ",humidity)
```

```python
visibility = current_weather['visibility']
print("Visibility: ",visibility)
```

```python
wind_speed = current_weather['wind']['speed']
print("Wind Speed: ",wind_speed)
```

# Air Pollution API

```python
url2="http://api.openweathermap.org/data/2.5/air_pollution?lat="+str(lat)+"&lon="+str(long)+"&units=metric&appid="+api_key
response2 = requests.get(url2)
air_pollution = response2.json()
# print(current_weather)
print(json.dumps(air_pollution, indent = 5))
```

```python
aqi = air_pollution['list'][0]['main']['aqi']
print(aqi)
```

```python
if(aqi==1):
    print("Good")
elif(aqi==2):
    print("Fair")
elif(aqi==3):
    print("Moderate")
elif(aqi==4):
    print("Poor")
elif(aqi==5):
    print("Very Poor")
```

```python
quality = {1:'Good', 2:'Fair', 3:'Moderate', 4:'Poor', 5:'Very Poor'}
print("Air Quality: ",quality[aqi])
```

# Call 5 day / 3 hour forecast data

```python
url3="http://api.openweathermap.org/data/2.5/forecast?lat="+str(lat)+"&lon="+str(long)+"&appid="+api_key
response3 = requests.get(url3)
fiveday = response3.json()
# print(current_weather)
print(json.dumps(fiveday, indent = 5))
```

```python
D = {"date_time":[], "temp":[], "pressure":[], "humidity":[], "weather":[]}

for i in fiveday['list']:
    D["date_time"].append(i["dt_txt"])
    D["temp"].append(i["main"]["temp"])
    D["pressure"].append(i["main"]["temp"])
    D["humidity"].append(i["main"]["humidity"])
    D["weather"].append(i["weather"][0]["description"])
```

```python
import pandas as pd
df=pd.DataFrame(D)
df
```
